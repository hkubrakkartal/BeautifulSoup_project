```python
import requests
from bs4 import BeautifulSoup
from googlesearch import search
import pandas as pd
import numpy as np
```

```python
def analysis(url):
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')
    print('H1 - H2 - H3 ANALİZ')
    
    h1 = soup.find_all("h1")
    print("Bu site içinde H1 tagı {} defa kullanılmıştır.".format(len(h1)))
    h2 = soup.find_all("h2")
    print("Bu site içinde H2 tagı {} defa kullanılmıştır.".format(len(h2)))
    h3 = soup.find_all("h3")
    print("Bu site içinde H3 tagı {} defa kullanılmıştır.".format(len(h3)))
    
    
    print('TİTLE ANALİZ')
    
    character = words = np.nan
    text = value = np.nan
    try:         
        title_tag = soup.title
        print(title_tag)
        words=0
        character = len(title_tag.string.strip())
        text = title_tag.string.strip()

        print("Title içinde {} karakter içermektedir".format(character))
        for word in title_tag.string.split():
            words = words+1
        print("Title içinde {} kelime vardır".format(words))
        if '<h1>' in title:
            print('Title içinde H1 tagı vardır.')
            value = 1
        else :
            print('Title içinde H1 tagı bulunmamaktadır.')
            value = 0
    except Exception as e:  
        print('Title bulunmamaktadır.',e) 
        
    return len(h1),len(h2),len(h3),text, character, words, value
```

```python
n = 15
query_lst = []
for i in range(15):
    q = input('Aramak yapmak için bir şeyler yazın: ')
    query_lst.append(q) 
print('aranıyor..')
c=0
results_list = []
data = pd.DataFrame(columns=["order","query","url","h1","h2","h3","title","title_words","title_ch","isH1"])
while True:
    
    order = 1
    for url in search(query_lst[c], tld = 'com', lang = 'tr', num = 15, start = 0,stop = 15):
        results_list.append(url)
        print('\n',url)
        print('analiz ediliyor..')
        
        h1,h2,h3,title, character,words, value = analysis(url)
        data = data.append({"order":order,"query":query_lst[c],"url":url,"h1":h1,"h2":h2,"h3":h3,
                            "title":title,"title_words":words,"title_ch":character,"isH1":value}, ignore_index=True)
        order = order+1
    print('\n*********************')
    c=c+1
    if c==15:
        break
        
print('---SON---')
```
